# Regression Analysis
Regression analysis is statistical method to model the relationship between a dependent and independent variable with one or more independent variables.

It helps us to understand how the value of dependent variable is changing corresponding to an independent variable when the independent variable is held fixed.

Regression analysis is used for one of two purposes: predicting the value of the dependent variable when information about the independent variables is known or predicting the effect of an independent variable on the dependent variable.

#### Dependent variables
Variables that depend on another variables, or the factors in regression analysis which we want to predict or understand are called <b>dependent variables</b>

#### Independent variables
the factors which affect the dependent variables that are used to predict thee values of dependent variable are called <b>Independent Variables</b>

#### Outlairs
An observation which contains either very low or very large values is called <b>Outlair</b>

## Linear Regression
Statistical method used for predective analysis.<br>
Linear regression assumes a linear connection between a dependent variable (Y) and an independent variable (X).<br>
It employs a regression line, also known as a best-fit line.<br> 
The linear connection is defined as Y = c+m*X + e, where c denotes the intercept, m denotes the slope of the line, and e is the error term.<br>
